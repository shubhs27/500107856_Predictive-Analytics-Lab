```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

img = mpimg.imread('cool bird.jpg')

fig, ax = plt.subplots(6, 10, figsize=(15, 10))
for i in range(6):
    for j in range(10):
        ax[i, j].imshow(img)
        ax[i, j].axis('off')
plt.suptitle('6x10 Grid of Images')
plt.show()

fig, ax = plt.subplots(10, 10, figsize=(15, 10))
for i in range(10):
    for j in range(10):
        if j == i or j == 9 - i:
            ax[i, j].imshow(img)
        ax[i, j].axis('off')
plt.suptitle('X Shape of Images')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('Pokemon.csv', encoding='ISO-8859-1')  # Adjust the file path if necessary

# Scatter plot of Attack vs Defense
plt.figure(figsize=(10, 6))
plt.scatter(df['Attack'], df['Defense'], alpha=0.7, c='blue', edgecolor='k')
plt.title('Scatter Plot of Attack vs Defense')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.grid(True)
plt.show()
```
